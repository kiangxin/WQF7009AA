```python
import pandas as pd

# Load dataset
csv_path = "air_quality_health_impact_data.csv"
df = pd.read_csv(csv_path)

df.head()
```

```python
# Column types and non-null counts
df.info()
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Summary statistics for numeric columns
numeric_summary = df.describe()
numeric_summary
```

```python
print("\nDuplicate Rows Count:", df.duplicated().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Modern-style target distribution plot
sns.set_theme(style="whitegrid", context="notebook")
order = df["HealthImpactClass"].value_counts().sort_index().index

plt.figure(figsize=(8, 5))
ax = sns.countplot(
    x="HealthImpactClass",
    data=df,
    order=order,
    palette="viridis",
)
ax.bar_label(ax.containers[0], fmt="%d")
plt.title("Distribution of Health Impact Classes")
plt.xlabel("Health Impact Class")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, fmt='0.2f', cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# AQI distribution by Health Impact Class
plt.figure(figsize=(10, 6))
ax = sns.kdeplot(
    data=df,
    x="AQI",
    hue="HealthImpactClass",
    fill=True,
    common_norm=False,
    alpha=0.35,
)
plt.title("AQI Distribution by Health Impact Class")
plt.xlabel("AQI")
plt.ylabel("Density")
plt.tight_layout()
plt.show()
```
